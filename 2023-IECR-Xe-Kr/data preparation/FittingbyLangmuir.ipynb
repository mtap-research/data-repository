```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import pandas as pd
```

## Define Langmuir model

```python
def langmuir(p,M,K):
    return M*K*p/(1+K*p)
```

```python
data = pd.read_excel("FitttingData.xlsx")
```

## Fitting

```python
MOF_names = data["Filename"]
```

```python
for name in MOF_names:
    MOF = str(name)
    data_Xe = pd.read_csv('./isotherms/'+MOF+'_Xe.csv')
    data_Kr = pd.read_csv('./isotherms/'+MOF+'_Kr.csv')

    gas_list = ["Xe", "Kr"]
    data_all = [data_Xe, data_Kr]
    
    for i in range(2):
        
        x = data_all[i]["p"]
        y = data_all[i]["q"]

        try:
            popt, pcov = curve_fit(langmuir, x, y,maxfev = 100000)

            M = popt[0] 
            K = popt[1]
            
            y_pre = langmuir(x,M,K)

            cor = np.corrcoef(y, y_pre)

            print(MOF,gas_list[i],M, K,cor[1][0])
        except:
            print(MOF,gas_list[i],"Fail fit")
```

## Plot

```python
fig = plt.figure(figsize=(13,5),dpi=600)

ax = fig.add_subplot(121)
bx = fig.add_subplot(122)


ax.plot([-10,300],[-10,300],'--',color='0.5',zorder=0)

ax.set_xscale('log')
ax.set_yscale('log')

X_mk1,Y_mk1 = data["Xe_M"].dropna(),data["Kr_M"].dropna()
ax1=ax.hexbin(X_mk1,Y_mk1, gridsize=(40,20), vmin=0,vmax=60,
              xscale='log',yscale='log',mincnt=0.1,edgecolors='None',cmap='viridis_r')

ax.set_xlim(0.1,300)
ax.set_ylim(0.1,300)

ax.set_xlabel(r'M of Xe / mmolg$^{-1}$',fontdict={'family':'Calibri','size':22})
ax.set_ylabel(r"M of Kr / mmolg$^{-1}$",fontdict={'family':'Calibri','size':22})

x=[0.1,1,10,100]

ax.set_xticks(x)
ax.set_yticks(x)
ax.spines['top'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['left'].set_linewidth(1.5)
ax.spines['right'].set_linewidth(1.5)
ax.tick_params(axis='both', which='both', direction='out',width=1.5,labelsize=16)
ax.minorticks_off()

ax.text(0.05,0.9,"a",fontdict={'family':'Calibri','size':24,'fontweight':'bold'},transform=ax.transAxes)


bx.plot([-10,300],[-10,300],'--',color='0.5',zorder=0)

bx.set_xscale('log')
bx.set_yscale('log')

X_mk2,Y_mk2 = data["Xe_K"].dropna(),data["Kr_K"].dropna()
bx1=bx.hexbin(X_mk2,Y_mk2, gridsize=(40,20), vmin=0,vmax=60,
              xscale='log',yscale='log', mincnt=0.1,edgecolors='None',cmap='viridis_r')

bx.set_xlim(0.001,300)
bx.set_ylim(0.001,300)

bx.set_xlabel(r'K of Xe / bar$^{-1}$',fontdict={'family':'Calibri','size':22})
bx.set_ylabel(r"K of Kr / bar$^{-1}$",fontdict={'family':'Calibri','size':22})

y=[0.001,0.01,0.1,1,10,100]
bx.set_xticks(y)
bx.set_yticks(y)
bx.spines['top'].set_linewidth(1.5)
bx.spines['bottom'].set_linewidth(1.5)
bx.spines['left'].set_linewidth(1.5)
bx.spines['right'].set_linewidth(1.5)
bx.tick_params(axis='both', which='both', direction='out',width=1.5,labelsize=16)
bx.minorticks_off()

bx.text(0.05,0.9,"b",fontdict={'family':'Calibri','size':24,'fontweight':'bold'},transform=bx.transAxes)

bx.text(1.33,0.5,"Counts",fontdict={'family':'Calibri','size':24},ha='center',va='center',rotation=90,transform=bx.transAxes)
cbar_ax = fig.add_axes([0.93, 0.1, 0.015, 0.8])
fig.colorbar(ax1, ax=[ax,bx], cax=cbar_ax)
cbar_ax.tick_params(labelsize=16)

plt.subplots_adjust(left=0.125,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.5, 
                    hspace=0.35)
plt.show()
```
