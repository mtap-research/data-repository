```python
from rdkit import Chem
from rdkit.Chem import Descriptors
from rdkit.ML.Descriptors import MoleculeDescriptors
import pandas as pd
```

```python
smis = pd.read_csv("ipcc_dataset.csv")["SMILE"]
```

```python
descs = [desc_name[0] for desc_name in Descriptors._descList]

desc_calc = MoleculeDescriptors.MolecularDescriptorCalculator(descs)
descriptors = pd.DataFrame([desc_calc.CalcDescriptors(Chem.MolFromSmiles(smi)) for smi in smis])
descriptors.columns = descs
descriptors.to_csv('all_rdkit_fea.csv',index=None)
```

```python
dft_features = pd.read_csv("ipcc_dataset.csv")[["IP","EA","HomoLumoGap"]].reset_index(drop=True)
target = pd.read_csv("ipcc_dataset.csv")[["LF","GWP"]].reset_index(drop=True)
descriptors = descriptors.reset_index(drop=True)
df_all_features = pd.concat([dft_features, descriptors, target], axis=1)
df_all_features.to_csv("data4pretrain.csv", index=None, index_label='Number')
```
